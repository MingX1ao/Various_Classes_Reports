## 1. 显示图像并调节对比度

### 1.1 显示图像和基本信息

```python
import pydicom as dcm
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds = dcm.dcmread("00000001.dcm")
pixel_array = ds.pixel_array

rows = ds.Rows
columns = ds.Columns
bitsStored = ds.BitsStored

print(f"图像行数: {rows}")
print(f"图像列数: {columns}")
print(f"存储的比特数: {bitsStored}")

grayscaleLevels = 2 ** bitsStored
print(f"灰度级: {grayscaleLevels}")

plt.figure(figsize=(5, 5))
plt.imshow(pixel_array, cmap=plt.cm.gray)
plt.title("An IMG")
plt.axis('off')  
```

图像的灰度级是65536，灰度比特是16，类型为uint16

### 1.2 调节对比度

将灰度级从65536变为256，降低对比度

```python
pixel_arrayChange = np.array((ds.pixel_array-np.min(ds.pixel_array)) * 256 / (np.max(ds.pixel_array)-np.min(ds.pixel_array)))
pixel_arrayChange = pixel_arrayChange.astype(int)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(pixel_array, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off') 

axes[1].imshow(pixel_arrayChange, cmap=plt.cm.gray)
axes[1].set_title('Low')
axes[1].axis('off') 

plt.tight_layout()
```

可以发现对比度变低了

## 2 对图像的操作

### 2.1 阈值变换

```python
yuzhi = 2 ** 6
yuzhiArray = np.empty((rows, columns))

for i in range(0, rows):
    for j in range(0, columns):
        if ds.pixel_array[i][j] < yuzhi:
            yuzhiArray[i][j] = 0
        else:
            yuzhiArray[i][j] = 2 ** 16 -1
                

            
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(pixel_array, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off') 

axes[1].imshow(yuzhiArray, cmap=plt.cm.gray)
axes[1].set_title('yuzhi')
axes[1].axis('off') 

plt.tight_layout()
```

### 2.2 图像反转

```python
inversArray = np.empty((rows, columns))

for i in range(0, rows):
    for j in range(0, columns):
        inversArray[i][j] = 2 ** 16 - ds.pixel_array[i][j] - 1

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(pixel_array, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off') 

axes[1].imshow(inversArray, cmap=plt.cm.gray)
axes[1].set_title('invers')
axes[1].axis('off') 

plt.tight_layout()
```

### 2.3 对数变换

```python
import math

logArray = np.empty((rows, columns))
c = 100

for i in range(0, rows):
    for j in range(0, columns):
        logArray[i][j] = c * math.log10(1+ds.pixel_array[i][j])

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(pixel_array, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off') 

axes[1].imshow(logArray, cmap=plt.cm.gray)
axes[1].set_title('log')
axes[1].axis('off') 

plt.tight_layout()
```

### 2.4 伽马变换

```python
gamma1Array = np.empty((rows, columns))
gamma2Array = np.empty((rows, columns))
c = 100
gamma1 = 0.5
gamma2 = 2


for i in range(0, rows):
    for j in range(0, columns):
        gamma1Array[i][j] = c * ds.pixel_array[i][j] ** gamma1
        gamma2Array[i][j] = c * ds.pixel_array[i][j] ** gamma2
        
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(pixel_array, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off') 

axes[1].imshow(gamma1Array, cmap=plt.cm.gray)
axes[1].set_title('gamma<1')
axes[1].axis('off') 

axes[2].imshow(gamma2Array, cmap=plt.cm.gray)
axes[2].set_title('gamma>1')
axes[2].axis('off') 


plt.tight_layout()
```

## 3 直方图均衡化 

由于原图的灰度级太多，先将灰度级降为256  
图像对比度不佳的原因：
1. 灰度太集中，范围不广，对比度低
2. 在低灰度处有一个很大的尖峰，分布不均匀，对比度低


```python
import seaborn as sns

L = 256
hist = np.zeros(L)
#画直方图
for i in range(0, rows):
    for j in range(0, columns):
        hist[pixel_arrayChange[i][j]] = hist[pixel_arrayChange[i][j]] + 1

gray = np.arange(L)

#直方图均衡化
tmp = 0
equalizedCheck = np.zeros(256)
for i in range(0,L):
    tmp += hist[i]
    equalizedCheck[i] = (L-1) / rows / columns * tmp
    
equalizedCheck = np.round(equalizedCheck).astype(int)


equalizedArray = np.empty((rows, columns))    
for i in range(0, rows):
    for j in range(0, columns):
        equalizedArray[i][j] = equalizedCheck[pixel_arrayChange[i][j]]
equalizedArray = equalizedArray.astype(int)
        
equalizedHist = np.zeros(L)        
for i in range(0, rows):
    for j in range(0, columns):
        equalizedHist[equalizedArray[i][j]] = equalizedHist[equalizedArray[i][j]] + 1        

        
        
        
fig1, axes1 = plt.subplots(1, 2, figsize=(10, 5))
axes1[0].imshow(pixel_arrayChange, cmap=plt.cm.gray)
axes1[0].set_title('Original')
axes1[0].axis('off') 

axes1[1].imshow(equalizedArray, cmap=plt.cm.gray)
axes1[1].set_title('Equalized')
axes1[1].axis('off') 


fig2, axes2 = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(x=gray, weights=hist[gray], bins=256, ax=axes2[0])
axes2[0].set_title('Original Histogram')

sns.histplot(x=gray, weights=equalizedHist[gray], bins=256, ax=axes2[1])
axes2[1].set_title('Equalized Histogram')
```

### 4.2 编程验证

```python
size = 8
pixArray = np.array([
    [2, 3, 2, 5, 1, 2, 3, 2],
    [4, 13, 14, 15, 10, 8, 14, 1],
    [3, 14, 8, 14, 9, 9, 15, 2],
    [5, 15, 13, 14, 8, 9, 14, 0],
    [6, 12, 10, 9, 6, 11, 13, 0],
    [4, 13, 5, 10, 7, 12, 15, 1],
    [7, 15, 12, 14, 4, 11, 14, 2],
    [1, 2, 4, 3, 6, 5, 3, 6]
])

gray = np.zeros(16)
x= np.arange(16)
for i in range(0,size):
    for j in range(0,size):
        gray[pixArray[i][j]] += 1
gray = gray.astype(int)

sns.histplot(x=x, weights=gray[x], bins=16)
```

```python
L = 16
tmp = 0
eqCheck = np.zeros(L)
for i in range(0,L):
    tmp += gray[i]
    eqCheck[i] = (L-1) * tmp / (size ** 2)
eqCheck = np.round(eqCheck).astype(int)
    
eqGray = np.zeros(L)
for i in range(0,L):
    eqGray[eqCheck[i]] += gray[i]
sns.histplot(x=x, weights=eqGray[x], bins=16)

eqGray
```
